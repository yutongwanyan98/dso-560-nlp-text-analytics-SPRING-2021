Write the code to Identify all dollar values ($2.99) and time references (11:43pm) using named capture groups in the poor amazon dataset.

```python
import pandas as pd
import re
reviews = open("poor_amazon_toy_reviews.txt").readlines()
```

```python
regex_pattern = r'(?P<dollar_sign>\$)(?P<dollar>\d+)\.?(?P<cents>\d+)?'
# this pattern includes all the $x.xx format and $x format;
# if it's $x, the cents will be 'None'

#file.seek(0) # reset the index position
#reviews = file.readlines()
x
for review in reviews:
    match = re.search(regex_pattern, review)
    if match:
        #if match.groupdict()['dollar_sign'] != None:
        print(match.groupdict()) 

# there are also some dollar values written in format "xx$" like 30$, 15$ etc., but I didnt include them in this particular pattern.
```
